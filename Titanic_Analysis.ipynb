```
def mr_mrs(name):
    if "Mr." in name:
        return "Mr"
    elif "Mrs." in name:
        return "Mrs"
    elif "Master." in name:
        return "Master"
    else:
        return "Ms"
```

```
tdf["Prefix"]= tdf['Name'].apply(mr_mrs)
tdf[tdf['Age']<10]
```

```
fig, (axis1, axis2)= plt.subplots(1,2, sharey= True)
sns.countplot('Pclass', data= tdf, hue='Prefix', ax= axis1)
sns.countplot('Survived', data= tdf, hue='Prefix', ax= axis2)
```

```
#Storing the survived people Data as a seperate Dataframe
tdf.head()
```

```
survived = tdf[tdf["Survived"] == 1]
```

```
#Replacing the Prefix of survived people
survived["Prefix"].replace("Mr", "Sur Mr", inplace = True)
survived["Prefix"].replace("Mrs", "Sur Mrs", inplace = True)
survived["Prefix"].replace("Ms", "Sur Ms", inplace = True)
survived["Prefix"].replace("Master", "Sur Master", inplace = True)
survived.head()
```
